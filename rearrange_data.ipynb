```python
#select the needed rows
odf=odf.loc[1:,"B001_01":"B020_05"]
odf
```

```python
#function to get gallery
def get_gallery(str):    
    return '-'
 
#function to get price
def get_price(str):
    return 0
```

```python
#function to transform df
def transform(odf):
    #get data head
    head = list(odf.columns.values)

    #create new d
    df = pd.DataFrame(columns=['Case','Pciture Series','Picture','Price','Gallery','Rank'])
    
    #iterate through every row
    for i in range(len(odf)):
        
        #iterate through every column
        for j in range(len(head)):
            
            column_head = head[j]
            
            #create new row
            cas = i
            pcs = int(column_head[2] + column_head[3])
            pct = int(column_head[6])
            prc = get_price(column_head)
            gll = get_gallery(column_head)
            rnk = odf.iloc[i, j]
            
            df2 = pd.DataFrame({'Case': [cas],
                    'Pciture Series' : [pcs],
                    'Picture' : [pct],
                    'Price' : [prc],
                    'Gallery' : [gll],
                    'Rank' : [rnk]})
            
            df = pd.concat([df, df2], ignore_index = True, axis = 0)
            
    return df
```

```python
#create new df
df = transform(odf)
df
```
