```python
# read orignal data
odf = pd.read_csv('edata.csv', encoding = "ISO-8859-1")

# read info data
idf = pd.read_csv('dataset\data-infos.csv', encoding = "ISO-8859-1", sep = ';')
```

```python
#select needed rows
odf=odf.loc[1:,"B001_01":"B020_05"]

# explore orignal data
odf
```

```python
#function to transform df
def transform(odf):
    #get data head
    head = list(odf.columns.values)

    #create new d
    df = pd.DataFrame(columns=['Case','Pciture Series','Picture','Price','Gallery','Rank'])
    
    #iterate through every row
    for i in range(len(odf)):
        
        #iterate through every column
        for j in range(len(head)):
            
            column_head = head[j] # column name: 'B0XX_0Y'
            
            # create new row
            cas = i #'Case': every row of odf is one case
            
            pcs = int(column_head[2] + column_head[3]) #'Pciture Series': get XX from 'B0XX_0Y'
            pct = int(column_head[6]) #'Picture': get Y from 'B0XX_0Y'
            
            prc = idf.iloc[j, 2] #'Price': get price from idf
            gll = idf.iloc[j, 3] #'Gallery': get gallery from idf
            
            rnk = odf.iloc[i, j] #'Rank': get rank from odf
            
            df2 = pd.DataFrame({'Case': [cas],
                    'Pciture Series' : [pcs],
                    'Picture' : [pct],
                    'Price' : [prc],
                    'Gallery' : [gll],
                    'Rank' : [rnk]})
            
            df = pd.concat([df, df2], ignore_index = True, axis = 0)
            
    return df
```

```python
#create new df
df = transform(odf)
df
```
