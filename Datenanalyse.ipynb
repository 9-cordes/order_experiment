```python
# read orignal data
odf = pd.read_csv('data.csv', encoding = "ISO-8859-1")

# read info data
idf = pd.read_csv('dataset\data-infos.csv', encoding = "ISO-8859-1", sep = ';')
```

```python
number_of_cases = len(odf)
```

```python
#function to transform df
def transform(odf):
    #get data head
    head = list(odf.columns.values)

    #create new d
    df = pd.DataFrame(columns=['Case','Picture Series','Picture','Price','Gallery','Category','Rank'])
    
    #iterate through every row
    for i in range(len(odf)):
        
        #iterate through every column
        for j in range(len(head)):
            
            column_head = head[j] # column name: 'B0XX_0Y'
            
            # create new row
            cas = i #'Case': every row of odf is one case
            
            pcs = int(column_head[2] + column_head[3]) - 1 #'Picture Series': get XX from 'B0XX_0Y'
            pct = int(column_head[6]) - 1 #'Picture': get Y from 'B0XX_0Y'
            
            prc = idf.iloc[j, 2] #'Price': get price from idf
            gll = idf.iloc[j, 3] #'Gallery': get gallery from idf
            ctg = idf.iloc[j, 4] #'Category': get category from idf
            
            rnk = odf.iloc[i, j] #'Rank': get rank from odf
            
            df2 = pd.DataFrame({'Case': [cas],
                    'Picture Series' : [pcs],
                    'Picture' : [pct],
                    'Price' : [prc],
                    'Gallery' : [gll],
                    'Category' : [ctg],            
                    'Rank' : [rnk]})
            
            df = pd.concat([df, df2], ignore_index = True, axis = 0)
            
    return df
```

```python
#create new df
df = transform(odf)

# explor new transformed data
pd.set_option('display.max_rows', None)
df
```

## 3. Statistische Analyse

### pearson correlation

#### correlation for each picture series --> fischer z transformation --> mean of Z --> transformation to correlation

```python
#convert ranks from String to int
df['Rank'] = df['Rank'].astype('int')
#calculate sums
df_sum = df.groupby(['Picture Series','Picture'],as_index=False).agg({'Rank': 'sum'})

# convert to np array
data = np.zeros((100, 2))

for i in range(100):
    #price
    data[i][0] = float(idf.iloc[i, 2])
    #rank
    data[i][1] = df_sum.iloc[i,2]

    
# calculate pearson correlation for each picture series
r = np.zeros((20))
for i in range(20):
    pos = i*5
    data_ps = data[pos:pos+5]
    
    x = data_ps[:,0]
    y = data_ps[:, 1]
    
    if np.amax(x)-np.amin(x) != 0:
        r[i] = np.corrcoef(x,y)[0][1]
    else:
        r[i] = 0

# fischer z transformation
Z = np.zeros((20))
for i in range(20):
    
    Z[i] = 0.5 * np.log((1+r[i]) / (1-r[i]))

# get mean 
Z_ = np.mean(Z)

# transfrom back to corr
r_ = (np.exp(2 * Z_) - 1) / (np.exp(2 * Z_) + 1)

r_
```

### spearman correlation

### pearson for each category

#### categroy 1

```python
#convert ranks from String to int
df['Rank'] = df['Rank'].astype('int')
#calculate sums
df_sum = df.groupby(['Picture Series','Picture'],as_index=False).agg({'Rank': 'sum'})

# convert to np array
data = np.zeros((75, 2))

for i in range(75):
    #price
    data[i][0] = float(idf.iloc[i, 2])
    #rank
    data[i][1] = df_sum.iloc[i,2]

    
# calculate pearson correlation for each picture series
r = np.zeros((15))
for i in range(15):
    pos = i*5
    data_ps = data[pos:pos+5]
    
    x = data_ps[:,0]
    y = data_ps[:, 1]
    
    if np.amax(x)-np.amin(x) != 0:
        r[i] = np.corrcoef(x,y)[0][1]
    else:
        r[i] = 0

# fischer z transformation
Z = np.zeros((15))
for i in range(15):
    
    Z[i] = 0.5 * np.log((1+r[i]) / (1-r[i]))

# get mean 
Z_ = np.mean(Z)

# transfrom back to corr
r_ = (np.exp(2 * Z_) - 1) / (np.exp(2 * Z_) + 1)

r_
```

```python
# is r_ statistically significant
# alpha value
alpha = 0.1
# t value
df_ = number_of_cases - 2
t_df = (r_ * np.sqrt(df_)) / (np.sqrt(1 - r_**2))

t_crit = scipy.stats.t.ppf(q=1-alpha/2,df = df_)

# if true then r_ is not significant
print(t_df <= t_crit)
print(t_df)
print(t_crit)
```
